```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
class Perceptron:
    def __init__(self, learning_rate=0.1, n_iters=10):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.weights = None
        self.bias = None
    def learn(self, X, y):
        n, m = X.shape
        # initialize parameters
        self.weights = np.random.uniform(m)
        self.bias = 0
        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                f_x = np.dot(x_i, self.weights) + self.bias
                y_pred = self.activation_function(f_x)
                delta = self.lr * (y[idx]-y_pred)
                self.weights += delta * x_i
                self.bias += delta
        return
    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = self.activation_func(linear_output)
        return y_predicted
    def activation_function(self, x):
        return np.where(x>=0, 1, 0)
```

```python
X, y = datasets.make_blobs(n_samples=1500, n_features=2,
                           centers=2, cluster_std=1.5,
                           # random_state=42
                          )
# X[:5], y[:5]
plt.scatter(X[:,0], X[:,1],
            c=y, marker='.'
           )
plt.plot(X[np.argmin(X[:,0])], X[np.argmax(X[:,0])])
plt.title('Before learning')
plt.savefig('before_learning.png')
plt.show()
```

```python
X.shape, y.shape
```

```python
model = Perceptron()
model.learn(X, y)

plt.scatter(X[:,0], X[:,1],
            marker=".", c=y
           )

x0_1 = np.amin(X[:, 0])
x0_2 = np.amax(X[:, 0])

x1_1 = (-model.weights[0]*x0_1 - model.bias) / model.weights[1]
x1_2 = (-model.weights[0]*x0_2 - model.bias) / model.weights[1]

ymin = np.amin(X[:, 1])
ymax = np.amax(X[:, 1])

plt.ylim([ymin-3, ymax+3])
plt.plot([x0_1, x0_2], [x1_1, x1_2])
plt.title('After learning')
plt.savefig('after_learning.png')
plt.show()
```
